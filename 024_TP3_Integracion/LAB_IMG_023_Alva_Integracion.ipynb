## Sección 1: Detección de Rostros con Haar Cascades

**Concepto:** Las Haar Cascades son un método rápido basado en características visuales simples (diferencias de intensidad en patrones tipo Haar) para detectar objetos, en este caso, rostros. El clasificador (`.xml`) contiene una "cascada" de filtros: la imagen pasa por filtros cada vez más específicos, y si falla en uno temprano, se descarta rápidamente, haciendo el proceso eficiente. Funciona mejor en escala de grises.

### 1.1 **Ejercicio 1:** Convertir la Imagen a RGB

OpenCV carga en BGR. Para mostrar con Matplotlib, la convertimos a RGB.

```
# Lectura de imagen usando OpenCV
img_bgr = cv2.imread('c7482082031c42fbba50975635525875.jpg') # Usamos el archivo que sabemos que existe

# Verificamos si la imagen se cargó correctamente
if img_bgr is None:
    print("Error: No se pudo cargar la imagen")
else:
    # Información básica de la imagen
    print("Dimensiones de la imagen:", img_bgr.shape)
    print("Tipo de datos:", img_bgr.dtype)
    print("Tamaño en memoria (bytes):", img_bgr.nbytes)

    # Convertimos de BGR a RGB para visualización correcta con matplotlib
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    
    # Visualización de la imagen completa
    plt.imshow(img_rgb)
    plt.title('Imagen en RGB')
    plt.axis('off')  # ON Muestra los ejes para referencia OFF No muestra los ejes
    #plt.colorbar()  # Añade barra de color pero no la muestra por estar comentada
    plt.show()
```

### 1.3 **Ejercicio 3:** Cargar Clasificador y Detectar Rostros

Cargamos el archivo `.xml` y usamos el método `detectMultiScale` sobre la imagen gris. Este método busca caras en diferentes tamaños y devuelve una lista de rectángulos `(x, y, ancho, alto)` por cada cara encontrada.

```
# Aquí le decimos a MediaPipe que use su "solución" para encontrar puntos faciales.
# mp.solutions.face_mesh contiene el modelo pre-entrenado.
# FaceMesh es la clase que usamos para crear el detector.
# static_image_mode=True: Le decimos que procese una imagen estática (no un video).
# max_num_faces=1: Le decimos que solo busque un rostro para simplificar.
# refine_landmarks=True: Intenta hacer los puntos un poco más precisos.
# min_detection_confidence=0.5: Solo considera un rostro si está seguro al menos en un 50%.
mp_face_mesh = mp.solutions.face_mesh.FaceMesh(
    static_image_mode=True,
    max_num_faces=2,
    refine_landmarks=True,
    min_detection_confidence=0.5)

print("Detector de puntos faciales inicializado.")

# Ahora procesamos la imagen. MediaPipe busca rostros y, si encuentra, sus puntos.
print("Procesando la imagen para encontrar puntos...")
resultados = mp_face_mesh.process(img_rgb)

# Una vez que terminamos de usar el detector, lo liberamos
mp_face_mesh.close()

print("Procesamiento completado.")
```

## Sección 2: Detección de Landmarks Faciales con LBF

**Concepto:** Una vez ubicada una cara (con su bounding box), podemos usar un detector de landmarks para encontrar puntos precisos en ella. El método LBF (Local Binary Features) usa un modelo (`lbfmodel.yaml`) entrenado para localizar 68 puntos estándar en el rostro. Necesita la imagen en escala de grises y las bounding boxes como entrada.

### 2.2 Procesar y Visualizar Landmarks (Pauls)

Los landmarks vienen en un formato específico (array 3D, tipo float). Los procesamos para obtener un array 2D de enteros `(x, y)` para cada uno de los 68 puntos del primer rostro (Pauls) y los dibujamos.

```
# Modificamos el código para usar resultados.multi_face_landmarks en lugar de all_landmarks
# ya que ese es el objeto que contiene los landmarks según el código anterior

# Procesar y dibujar landmarks
if resultados.multi_face_landmarks and len(resultados.multi_face_landmarks) >= 1:
    # Crear una copia de la imagen para dibujar
    img_con_landmarks = img_rgb.copy()
    
    # Procesar landmarks del primer rostro
    rostro = resultados.multi_face_landmarks[0]
    
    print("Dibujando landmarks de la Nieta (blanco):")
    # Dibujar cada punto landmark
    for punto in rostro.landmark:
        # Convertir coordenadas normalizadas a píxeles
        x = int(punto.x * ancho)
        y = int(punto.y * alto)
        cv2.circle(img_con_landmarks, (x, y), 1, (255, 255, 255), -1)  # Puntos blancos
    
    # Mostrar imagen completa con landmarks
    plt.figure(figsize=(12,8))
    plt.imshow(img_con_landmarks)
    plt.title("Rostros + Landmarks Nieta")
    plt.axis('off')
    plt.show()
    
    # Mostrar zoom (ajustamos las coordenadas según la posición del rostro)
    print("\nZoom en Nieta")
    # Calculamos las coordenadas para el zoom
    x_min = int(min(punto.x * ancho for punto in rostro.landmark))
    y_min = int(min(punto.y * alto for punto in rostro.landmark))
    x_max = int(max(punto.x * ancho for punto in rostro.landmark))
    y_max = int(max(punto.y * alto for punto in rostro.landmark))
    
    # Agregamos un margen al zoom
    margen = 20
    y1 = max(0, y_min - margen)
    y2 = min(alto, y_max + margen)
    x1 = max(0, x_min - margen)
    x2 = min(ancho, x_max + margen)
    
    plt.figure(figsize=(8,8))
    plt.imshow(img_con_landmarks[y1:y2, x1:x2])
    plt.title("Zoom Landmarks Nieta")
    plt.axis('off')
    plt.show()
else:
    print("No hay landmarks disponibles para procesar.")
```

### 3.1 **Ejercicio 6:** Dibujar Ojos y Boca (Nieta)

**Tarea:**
1.  Define listas con los índices correspondientes a los ojos y la boca.
2.  Crea una **copia limpia** de la imagen RGB original (`img_con_partes = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)`).
3.  Itera sobre `landmarks_nieta`.

5.  Muestra un zoom en Pauls de la imagen resultante.

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Índices según el modelo FaceMesh de MediaPipe (468 puntos)
indice_ojos_izquierdo = [33, 160, 158, 133, 153, 144, 145, 153]  # Ojo izquierdo
indice_ojos_derecho = [362, 385, 387, 263, 373, 380, 381, 373]   # Ojo derecho
indice_boca = [78, 191, 80, 81, 82, 13, 312, 311, 310, 415]  # Contorno de la boca

# Crear una copia limpia de la imagen RGB
img_con_partes = img_rgb.copy()

# Obtener landmarks de la imagen
if resultados.multi_face_landmarks:
    rostro = resultados.multi_face_landmarks[0]  # Primer rostro detectado
    alto, ancho, _ = img_rgb.shape

    # Convertir todos los landmarks del rostro en coordenadas (x, y)
    landmarks_nieta = [(punto.x * ancho, punto.y * alto) for punto in rostro.landmark]

    # Dibujar los puntos de los ojos izquierdo y derecho (magenta)
    for idx in indice_ojos_izquierdo + indice_ojos_derecho:
        x, y = landmarks_nieta[idx]
        cv2.circle(img_con_partes, (int(x), int(y)), 3, (255, 0, 255), -1)  # Magenta

    # Dibujar los puntos de la boca (cian)
    for idx in indice_boca:
        x, y = landmarks_nieta[idx]
        cv2.circle(img_con_partes, (int(x), int(y)), 3, (0, 255, 255), -1)  # Cian

    # Zoom automático en la región de los landmarks
    x_min = int(min(x for x, y in landmarks_nieta))
    x_max = int(max(x for x, y in landmarks_nieta))
    y_min = int(min(y for x, y in landmarks_nieta))
    y_max = int(max(y for x, y in landmarks_nieta))
    
    # Ajustamos el margen para evitar que se recorte la cara
    margen = 50
    x1 = max(0, x_min - margen)
    x2 = min(ancho, x_max + margen)
    y1 = max(0, y_min - margen)
    y2 = min(alto, y_max + margen)

    # Extraer la imagen con zoom
    zoom_img = img_con_partes[y1:y2, x1:x2]

    # Mostrar la imagen con zoom
    plt.figure(figsize=(6, 6))
    plt.imshow(zoom_img)
    plt.axis('off')
    plt.title("Zoom ojos y boca - Nieta")
    plt.show()
else:
    print("No se detectaron landmarks.")


```
